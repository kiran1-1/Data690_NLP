### Build 'word frequency'. Make sure you have removed the stopwords.
### Determine the maximum frequency as 'word_frequencies[word]=
### (word_frequencies[word]/maximum_frequency).

Here i have used  ‘spaceX_DP.txt’ file for word frequncy and for removing stop words i have used spacy

```python
import spacy
from collections import Counter

# Load the English tokenizer model
nlp = spacy.load("en_core_web_sm")

# Read the text file
with open('spaceX_DP.txt', 'r') as file:
    corpus = file.read()

# Process the text using spaCy
doc = nlp(corpus)

# Tokenize words, filter out stop words and non-alphabetic tokens, and convert to lowercase
words = [token.text.lower() for token in doc if token.is_alpha and not token.is_stop]

# Count word frequencies
word_freq = Counter(words)

# Print the most common 10 words and their frequencies
print("Word Frequency:")
for word, freq in word_freq.most_common(10):
    print(f"{word}: {freq}")

# Find maximum word frequency
maximum_frequency = max(word_freq.values())

# Normalize word frequencies
word_frequencies = {}
for word, freq in word_freq.items():
    word_frequencies[word] = freq / maximum_frequency

# Print the normalized word frequencies
print("Normalized Word Frequencies:")
for word, freq in word_frequencies.items():
    print(f"{word}: {freq}")
```

### Tokenize the sentences. Generate the sentence_scores. Score every sentence based on number of words.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')

# Read the text file
with open('spaceX_DP.txt', 'r') as file:
    corpus = file.read()  

# Tokenize sentences
sentences = sent_tokenize(corpus)

# Tokenize words
words = word_tokenize(corpus)

# Print tokenized sentences
print("Tokenized Sentences:")
for i, sentence in enumerate(sentences, 1):
    print(f"Sentence {i}: {sentence}")  # Print each sentence with its index

# Print tokenized words
print("\nTokenized Words:")
print(words)  # Print the list of tokenized words

# Calculate sentence scores based on number of words
sentence_scores = {}
for i, sentence in enumerate(sentences, 1):
    words = word_tokenize(sentence)
    num_words = len(words)
    sentence_scores[sentence] = num_words

# Print sentence scores (number of words)
print("Sentence Scores (Number of Words):")
for sentence, score in sentence_scores.items():
    print(f"{sentence}: {score}")
```

###  Import nlargest from heapq and provide 'summarized_sentences'.

```python
from heapq import nlargest

# Number of sentences to include in the summary
num_sentences = 5

# Get the top 5 sentences based on their scores
summarized_sentences = nlargest(num_sentences, sentence_scores, key=sentence_scores.get)

# Print the summarized sentences
print("Summarized Sentences:")
for sentence in summarized_sentences:
    print(sentence, "\n")
```

```python
# Define a function to preprocess the text
def preprocess_text(text):
    # Tokenize the text using spaCy, convert tokens to lowercase, remove stopwords, and filter out non-alphabetic tokens
    return ' '.join(token.text.lower() for token in nlp(text) if token.text.lower() not in STOP_WORDS and token.text.isalpha())

# Preprocess the corpus
preprocessed_text = preprocess_text(corpus)

# Count word frequencies
word_freq = Counter(preprocessed_text.split())

# Find the maximum word frequency
maximum_frequency = max(word_freq.values())

# Normalize word frequencies
word_frequencies = {word: freq / maximum_frequency for word, freq in word_freq.items()}

# Score sentences based on word frequencies
sentence_scores = {sentence: sum(word_frequencies[word] for word in preprocess_text(str(sentence)).split()) for sentence in doc.sents}

# Select top sentences based on scores
top_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:]

# Join top sentences to form the summary
summary = ' '.join(str(sentence) for sentence in top_sentences)

# Print the summary
print("Summary:")
print(summary)
```

###  Print the summary in 5 sentences.

```python
# Select the top 5 sentences based on their scores
top_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:5]

# Join the top sentences into a single string with newline characters separating them
summary = '\n'.join(str(sentence) for sentence in top_sentences)

# Print the summary with a header indicating that it contains 5 sentences
print("Summary (5 sentences):")
print(summary)
```

### Exercise 2. Sentiment Analysis

    What is the overall sentiment of the ‘spaceX.txt’ text? Provide the compound, negative,
    neutral, and positive metrics. Is the text more ‘subjective’ or more ‘objective’?

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# Calculate sentiment scores using SentimentIntensityAnalyzer
sentiment_scores = sia.polarity_scores(corpus_X)

# Print the sentiment scores
print("Sentiment Scores:")
print(sentiment_scores)

# Determine overall sentiment
if sentiment_scores['compound'] >= 0.05:
    print("Overall sentiment: Positive")
elif sentiment_scores['compound'] <= -0.05:
    print("Overall sentiment: Negative")
else:
    print("Overall sentiment: Neutral")

# Determine subjectivity of the text
if sentiment_scores['compound'] > 0.1:
    print("The text is more subjective.")
elif sentiment_scores['compound'] < -0.1:
    print("The text is more objective.")
else:
    print("The text has a balanced mix of subjective and objective content.")
```
