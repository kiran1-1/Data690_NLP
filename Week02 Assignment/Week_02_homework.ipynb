```python
from bs4 import BeautifulSoup
```

# 3.Connect to the website using urllib.request

```python
import urllib.request

url = "https://pythonprogramming.net/parsememcparseface/"  # Replace this with the URL of the website you want to connect to

with urllib.request.urlopen(url) as response:# Open the URL and read the HTML content
    html_content = response.read()

print(html_content.decode('utf-8'))
```

# 4.Create an object called âsoupâ using the source

```python

soup = BeautifulSoup(html_content, 'html.parser')
```

# 5.Print page title, get attributes, values, and beginning navigation. Get specific values.

```python

print("Page Title:", soup.title.text)

all_info = soup.find_all()
for X in all_info:
    for attribute, value in X.attrs.items():
        print(f"{attribute}: {value}")
```

```python
#Print the beginning of the navigation section (e.g., the first 'nav' tag)
first_nav_tag = soup.find('nav')
print("\nBeginning of Navigation Section:",first_nav_tag.text)
```

# 6. Find all the paragraph instead of just one in the previous query
# 7. You can iterate through them using âstr(paragraph.text)â

```python
all_paragraphs = soup.find_all('p')
print("\nAll Paragraphs:")
for paragraph in all_paragraphs:
    print(str(paragraph.text))
```

# 8. Grab all the links through âurl.get(âhere)

```python
all_links = [a.get('href') for a in soup.find_all('a')]
print("\nAll Links:")
for link in all_links:
    print(link)
```

# 9. Just grab the text using âget_textâ

```python
all_text = soup.get_text()
print("\nAll Text:")
print(all_text)
```
