```python
import string

file_path = '/Users/kiran_g/Documents/3rd sem/NLP/sentiment labelled sentences/amazon_cells_labelled.txt'

reviews = []  # To store the reviews
labels = []   # To store the corresponding labels
with open(file_path, 'r') as file:# Open the file in read mode ('r')
    for line in file:    # Read the file line by line
        review, label = line.strip().split('\t')# Split the line into review and label
        review = review.translate(str.maketrans('', '', string.punctuation)).lower()  # Remove punctuation from the review and convert to lowercase
        reviews.append(review)# Append the processed review and label to their respective lists
        labels.append(int(label))  # Convert label to integer
for i in range(5):# Display the first few reviews and labels to verify
    print("Review:", reviews[i])
    print("Label:", labels[i])
    print()
```

```python
vocabulary = set()  # Set to store unique words
for review in reviews:# Iterate over each review in the 'reviews' list
    review = review.translate(str.maketrans('', '', string.punctuation)).lower()    # Remove punctuation from the review and convert to lowercase
    words = review.split()# Tokenize the review into words
    vocabulary.update(words) # Add unique words to the vocabulary set
print("Vocabulary:", vocabulary)# Display the vocabulary
```

```python
word_to_int = {}  # Dictionary to store word-to-integer mapping
current_int = 0   # Start assigning integers from 0
for word in vocabulary:# Iterate over each word in the 'vocabulary' set
    word_to_int[word] = current_int    # Map the word to the current integer and increment the integer
    current_int += 1

print("Word-to-Integer Mapping:", word_to_int)
```

```python
encoded_reviews = []  # List to store encoded reviews
for review in reviews:
    review = review.translate(str.maketrans('', '', string.punctuation)).lower() # Remove punctuation from the review and convert to lowercase
    words = review.split() # Tokenize the review into words
    encoded_review = [word_to_int[word] for word in words if word in word_to_int] # Encode each word with its corresponding integer value
    encoded_reviews.append(encoded_review) # Append the encoded review to the list of encoded reviews

for i in range(len(reviews)):# Display the encoded reviews
    print("Review:", reviews[i])
    print("Encoded Review:", encoded_reviews[i])
    print()
```

```python
import torch
import torch.nn as nn

class MyLSTM(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_size):
        super(MyLSTM, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=3, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_size)
        
    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, _ = self.lstm(embedded)
        out = self.fc(lstm_out[:, -1, :])  # Use the output of the last time step
        return out
```
